```python
import numpy as np
import pylab as pl
import warnings

from matplotlib import pyplot as plt
from scipy import linalg
from sklearn.datasets import load_digits
from sklearn.decomposition import TruncatedSVD
from sklearn.ensemble import RandomForestClassifier
warnings.filterwarnings("ignore")
```

```python
from PIL import Image
img = Image.open("tomo1.jpg")
arr = np.array(img.getdata())
img2 = img.convert(mode='L', colors=256)
arr2 = np.array(img2.getdata())
X = arr2.reshape(512,512)
```

```python
pl.imshow(X)
```

```python
U, d, Vt = linalg.svd(X)
D = linalg.diagsvd(d, X.shape[0], X.shape[1])
```

```python
for k in range(10, 200, 20):
    D1 = D.copy()
    D1[D1 < d[int(k)]] = 0
    print (int(k))
    X1 = np.dot(np.dot(U, D1), Vt)
    pl.imshow(X1, cmap='gray')
    pl.show()
```
